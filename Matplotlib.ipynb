```python
catagory=["A", "B", "C", "D"]
value=[12, 4, 9, 7]
fig,  axes=plt.subplots(figsize=(6,3))
axes.bar(catagory,value,color='r')
plt.title("Bar chart")
plt.show()
```

```python
x = ["Mon", "Tue", "Wed", "Thu", "Fri"]
y = [5, 3, 6, 8, 2]
x2 = ["Red", "Blue", "Green"]
y2 = [10, 6, 9]
colors=['r','b','g']
fig, axes = plt.subplots(1, 2, figsize=(10,4))
axes[0].bar(x,y,color='g')
axes[1].barh(x2,y2,color=colors)
plt.show()
```
