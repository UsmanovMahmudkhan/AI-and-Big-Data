```python
def stemAndLeaf(data):
    stems = {}
    for num in data:
        stem = int(num // 10)
        leaf = int(num % 10)
        if stem not in stems:
            stems[stem] = []
        stems[stem].append(leaf)
    for stem, leaves in sorted(stems.items()):
        print(f"{stem}|{' '.join(map(str, sorted(leaves)))}")

data = [93, 17, 56, 42, 68, 91, 39, 58, 24, 83, 77, 65, 81, 50, 34, 20, 72, 65, 18, 61, 47, 88, 56, 33, 90]
stemAndLeaf(data)
```

```python
def symmetric_difference_recursive(sets):
  if len(sets) == 0:
    return set()
  if len(sets) == 1:
    return sets[0]

  result = sets[0]
  for i in range(1, len(sets)):
    result = result.symmetric_difference(sets[i])
  return result

set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = {5, 6, 7}
sets=[set1,set2,set3]
symmetric_diff = symmetric_difference_recursive(sets)
print(f"The symmetric difference of the given sets is: {symmetric_diff}")
```

```python
def difference(*sets):
  result=sets[0]
  for s in sets[1:0]:
    result=result^s
  return result

set1 = {1, 2, 3}
set2 = {3, 4, 5}
set3 = {5, 6, 7}
print(difference(set1,set2,set3))
```

```python
def generate_subsets(s):
    n = len(s)
    subsets = []
    for i in range(2 ** n):
        subset = [s[j] for j in range(n) if (i & (1 << j)) > 0]
        subsets.append(subset)
    return subsets
print(generate_subsets(['A', 'B', 'C']))
```

```python
def heaps_permutation(k, A):
    if k == 1:
        print(A)
    else:
        for i in range(k):
            heaps_permutation(k-1, A)
            if k % 2 == 0:
                A[i], A[k-1] = A[k-1], A[i]
            else:
                A[0], A[k-1] = A[k-1], A[0]
heaps_permutation(3, [1, 2, 3])
```

```python
p_values = [0.2, 0.5, 0.8]
x = [0, 1]

plt.figure(figsize=(3,3))
for p_val in p_values:
    pmf = [1-p_val, p_val]
    plt.plot(x, pmf, 'o-', label=f'p={p_val}', markersize=10)

plt.title('PMF of Bernoulli Distribution')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.xticks(x)
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```

```python
x_values = [0, 1, 2, 3, 4, 5, 6]
pmf = [0.05, 0.10, 0.15, 0.25, 0.20, 0.15, 0.10]
plt.bar(x_values, pmf)
plt.title('PMF of Computers in Use')
plt.show()
```

```python
cdf = np.cumsum(pmf)
plt.step(x_values, cdf, where='post')
plt.title('CDF of Computers in Use')
plt.show()
```

```python
expected_value = sum(x*p for x, p in zip(x_values, pmf))
print(f"Expected value: {expected_value}")
```

```python
variance = sum((x - expected_value)**2 * p for x, p in zip(x_values, pmf))
print(f"Variance: {variance}")
```

```python
std_dev = np.sqrt(variance)
print(f"Standard deviation: {std_dev}")
```

```python
k = 5  # number of successes
p = 0.3
neg_binom_samples = []
while len(neg_binom_samples) < 1000:
    trials = 0
    successes = 0
    while successes < k:
        trials += 1
        if np.random.random() < p:
            successes += 1
    neg_binom_samples.append(trials)
plt.hist(neg_binom_samples, bins=range(k, max(neg_binom_samples)+1))
plt.title('Negative Binomial Samples')
plt.show()
```
