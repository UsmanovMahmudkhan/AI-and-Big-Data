```python
x = np.array([1, 2, 3])
y = np.array([4, 5, 6])
joint_pmf = np.array([[0.1, 0.2, 0.1], [0.1, 0.2, 0.1], [0.05, 0.1, 0.05]])
cov = np.sum((x[:, None] - np.sum(x * joint_pmf.sum(axis=1))) * \
      (y - np.sum(y * joint_pmf.sum(axis=0))) * joint_pmf)

print(cov)
```

```python
cov = 0.5
sigma_x = 1.0
sigma_y = 1.0
corr = cov / (sigma_x * sigma_y)

print(corr)
```
