```python
import heapq

class Node:
    def __init__(self,name,cost):
        self.name=name
        self.cost=cost
    def __lt__(self,other):
        return self.cost<other.cost
def uniform_cost_search(graph,start,goal):
  visited=set()
  priority_queue=[(0,Node(start,0))]
  while priority_queue:
    cost,current_node=heapq.heappop(priority_queue)
    if current_node.name in visited:
      continue
    visited.add(current_node.name)
    if current_node.name==goal:
      return cost
    for neighbor,neighbor_cost in graph[current_node.name]:
      if neighbor not in visited:
        heapq.heappush(priority_queue,(cost+neighbor_cost,Node(neighbor,cost+neighbor_cost)))
  return float('inf')

graph={
    'S': [('A', 1), ('B', 4)],
    'A': [('C', 2), ('D', 5)],
    'B': [('D', 4)],
    'C': [('G', 3)],
    'D': [('G',5)]
}
start='S'
goal='G'
print("Uniform Cost Search")
print(uniform_cost_search(graph,start,goal))
```

```python
def greedy_coin_change(amount, coins):

    num_coins = 0
    remaining_amount = amount
    for coin in sorted(coins, reverse=True):
        while remaining_amount >= coin:
            remaining_amount -= coin
            num_coins += 1
    return num_coins

amount = 63
coins = [25, 10, 5, 1]
num_coins = greedy_coin_change(amount, coins)
print(f"Number of coins needed: {num_coins}")
```

```python
def greedy(graph, H_table, start, goal):
    visited = set()
    path = []
    current_node = start

    while current_node != goal:
        path.append(current_node)
        visited.add(current_node)

        neighbors = graph.get(current_node, [])
        if not neighbors:
            return None

        next_node = min(neighbors, key=lambda x: H_table[x[0]])

        current_node = next_node[0]

        if current_node in visited:
            return None
    path.append(goal)
    return path

graph = {
    'S': [('A', 3), ('B', 2)],
    'A': [('C', 1)],
    'B': [('D', 2)],
    'C': [('G', 3)],
    'D': [('G', 3)]
}

H_table = {
    'S': 10,
    'A': 8,
    'B': 5,
    'C': 5,
    'D': 4,
    'G': 0
}


path = greedy(graph, H_table, 'S', 'G')

if path:
    print("Path:", ' -> '.join(path))
else:
    print("No path found")
```

```python
import heapq

def astar(graph, H_table, start, goal):
    open_list = [(H_table[start], start, [])]
    closed_set = set()

    while open_list:
        f, current_node, path = heapq.heappop(open_list)

        if current_node == goal:
            return path + [current_node]

        closed_set.add(current_node)

        for neighbor, cost in graph.get(current_node, []):
            if neighbor not in closed_set:
                g = f - H_table[current_node] + cost
                heapq.heappush(open_list, (g + H_table[neighbor], neighbor, path + [current_node]))

    return None


graph = {
    'S': [('A', 3), ('B', 2)],
    'A': [('C', 1)],
    'B': [('D', 2)],
    'C': [('G', 3)],
    'D': [('G', 3)]
}

H_table = {
    'S': 10,
    'A': 8,
    'B': 5,
    'C': 5,
    'D': 4,
    'G': 0
}


path = astar(graph, H_table, 'S', 'G')

if path:
    print("Path:", ' -> '.join(path))
else:
    print("No path found")
```

```python
def evaluate(board):
    for row in range(3):
        if board[row][0] == board[row][1] == board[row][2]:
            return 10 if board[row][0] == 'X' else -10
    for col in range(3):
        if board[0][col] == board[1][col] == board[2][col]:
            return 10 if board[0][col] == 'X' else -10
    if board[0][0] == board[1][1] == board[2][2]:
        return 10 if board[0][0] == 'X' else -10
    if board[0][2] == board[1][1] == board[2][0]:
        return 10 if board[0][2] == 'X' else -10
    return 0

def is_moves_left(board):
    for row in board:
        if '_' in row:
            return True
    return False

def minimax(board, depth, is_max):
    score = evaluate(board)
    if score == 10 or score == -10:
        return score
    if not is_moves_left(board):
        return 0
    if is_max:
        best = -1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'X'
                    best = max(best, minimax(board, depth + 1, not is_max))
                    board[i][j] = '_'
        return best
    else:
        best = 1000
        for i in range(3):
            for j in range(3):
                if board[i][j] == '_':
                    board[i][j] = 'O'
                    best = min(best, minimax(board, depth + 1, not is_max))
                    board[i][j] = '_'
        return best

def find_best_move(board):
    best_val = -1000
    best_move = (-1, -1)
    for i in range(3):
        for j in range(3):
            if board[i][j] == '_':
                board[i][j] = 'X'
                move_val = minimax(board, 0, False)
                board[i][j] = '_'
                if move_val > best_val:
                    best_move = (i, j)
                    best_val = move_val
    return best_move

board = [
    ['X', 'O', 'X'],
    ['_', 'O', '_'],
    ['_', '_', '_']
]

best_move = find_best_move(board)
print(f"The best move is at position: {best_move}")
```

```python
def is_terminal(piles):
    return all(p == 0 for p in piles)

def evaluate(piles):
    return 1 if sum(piles) % 2 == 0 else -1

def minimax(piles, is_max):
    if is_terminal(piles):
        return evaluate(piles)

    if is_max:
        best = -float('inf')
        for i in range(len(piles)):
            for k in range(1, piles[i] + 1):
                new_piles = piles[:]
                new_piles[i] -= k
                best = max(best, minimax(new_piles, not is_max))
        return best
    else:
        best = float('inf')
        for i in range(len(piles)):
            for k in range(1, piles[i] + 1):
                new_piles = piles[:]
                new_piles[i] -= k
                best = min(best, minimax(new_piles, not is_max))
        return best

def find_best_move(piles):
    best_val = -float('inf')
    best_move = (-1, -1)
    for i in range(len(piles)):
        for k in range(1, piles[i] + 1):
            new_piles = piles[:]
            new_piles[i] -= k
            move_val = minimax(new_piles, False)
            if move_val > best_val:
                best_move = (i, k)
                best_val = move_val
    return best_move

piles = [3, 4, 5]
best_move = find_best_move(piles)
print(f"The best move is to remove {best_move[1]} stones from pile {best_move[0]}")
```

```python
#Second Example
import heapq

class Node:
    def __init__(self, name, heuristic):
        self.name = name
        self.heuristic = heuristic
        self.adjacency_list = []
        self.g = float('inf')
        self.f = float('inf')
        self.parent = None

    def __lt__(self, other):
        return self.f < other.f

def a_star(start_node, goal_node):
    open_list = []
    closed_list = set()
    heapq.heappush(open_list, start_node)
    start_node.g = 0
    start_node.f = start_node.heuristic

    while open_list:
        current_node = heapq.heappop(open_list)
        if current_node == goal_node:
            path = []
            while current_node:
                path.append(current_node.name)
                current_node = current_node.parent
            return path[::-1]

        closed_list.add(current_node)
        for neighbor, distance in current_node.adjacency_list:
            if neighbor in closed_list:
                continue

            tentative_g = current_node.g + distance
            if tentative_g < neighbor.g:
                neighbor.parent = current_node
                neighbor.g = tentative_g
                neighbor.f = neighbor.g + neighbor.heuristic

                if neighbor not in open_list:
                    heapq.heappush(open_list, neighbor)

    return None

new_york = Node('New York', 0)
boston = Node('Boston', 215)
washington = Node('Washington', 225)
philadelphia = Node('Philadelphia', 95)
pittsburgh = Node('Pittsburgh', 320)
cleveland = Node('Cleveland', 460)
chicago = Node('Chicago', 790)
detroit = Node('Detroit', 620)
miami = Node('Miami', 1300)
atlanta = Node('Atlanta', 860)

new_york.adjacency_list = [(boston, 215), (philadelphia, 95)]
boston.adjacency_list = [(new_york, 215)]
philadelphia.adjacency_list = [(new_york, 95), (washington, 140)]
washington.adjacency_list = [(philadelphia, 140), (pittsburgh, 245)]
pittsburgh.adjacency_list = [(washington, 245), (cleveland, 135), (detroit, 225)]
cleveland.adjacency_list = [(pittsburgh, 135), (chicago, 345)]
chicago.adjacency_list = [(cleveland, 345), (detroit, 280)]
detroit.adjacency_list = [(pittsburgh, 225), (chicago, 280)]
miami.adjacency_list = [(atlanta, 660)]
atlanta.adjacency_list = [(miami, 660), (washington, 640)]

path = a_star(new_york, chicago)
print("Path from New York to Chicago:", path)
```

```python
import numpy as np
global N
N = 8

def showResult(board):
    for i in range(N):
        for j in range(N):
            print(int(board[i][j]), end=' ')
        print()

def placement_safety(board, row, col):
    for i in range(col):
        if board[row][i] == 1:
            return False
    for i, j in zip(range(row, -1, -1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    for i, j in zip(range(row, N, 1), range(col, -1, -1)):
        if board[i][j] == 1:
            return False
    return True

def solution(board, col):
  if col>=N:
    return True
  for i in range(N):
    if placement_safety(board,i,col):
      board[i][col]=0
  return False

def main_NQ():
  board=np.zeros((8,8))
  if solution(board,0)==False:
    print("No solution!")
    return False
  showResult(board)

main_NQ()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def gaussian_mf(x, mean, sigma):
    return np.exp(-0.5 * ((x - mean) / sigma) ** 2)

x = np.linspace(0, 10, 100)
mean = 5
sigma_umf = 1.5
sigma_lmf = 2.0

umf = gaussian_mf(x, mean, sigma_umf)
lmf = gaussian_mf(x, mean, sigma_lmf)

plt.plot(x, umf, 'r--', label='Upper Membership Function (UMF)')
plt.plot(x, lmf, 'b--', label='Lower Membership Function (LMF)')
plt.fill_between(x, lmf, umf, color='gray', alpha=0.3, label='FOU (Footprint of Uncertainty)')

plt.xlabel('x')
plt.ylabel('Membership Degree')
plt.title('Type-2 Fuzzy Membership Function (Gaussian)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

data, labels = make_blobs(n_samples=500, centers=4, cluster_std=1.2, random_state=42)

def plot_clusters(data, labels, title):
    plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50)
    plt.title(title)
    plt.show()

kmeans = KMeans(n_clusters=4, random_state=42)
kmeans_labels = kmeans.fit_predict(data)
plot_clusters(data, kmeans_labels, "K-Means Clustering")

hierarchical = AgglomerativeClustering(n_clusters=4)
hierarchical_labels = hierarchical.fit_predict(data)
plot_clusters(data, hierarchical_labels, "Hierarchical Clustering")

dbscan = DBSCAN(eps=1.2, min_samples=5)
dbscan_labels = dbscan.fit_predict(data)
plot_clusters(data, dbscan_labels, "DBSCAN Clustering")

pca = PCA(n_components=2)
data_reduced = pca.fit_transform(data)
plot_clusters(data_reduced, labels, "PCA Reduced Data")

gmm = GaussianMixture(n_components=4, random_state=42)
gmm_labels = gmm.fit_predict(data)
plot_clusters(data, gmm_labels, "Gaussian Mixture Model")
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import matplotlib.pyplot as plt
import numpy as np

(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data("")
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)
y_train = keras.utils.to_categorical(y_train, 10)
y_test = keras.utils.to_categorical(y_test, 10)

def create_model():
    model = keras.Sequential([
        layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(10, activation='softmax')
    ])
    return model

def learning_rate_techniques():
    results = {}

    def step_decay(epoch):
        initial_lr = 0.001
        drop = 0.5
        epochs_drop = 10.0
        lr = initial_lr * np.power(drop, np.floor((1+epoch)/epochs_drop))
        return lr

    techniques = {
        'Step Decay': keras.callbacks.LearningRateScheduler(step_decay),
        'Exponential Decay': keras.callbacks.LearningRateScheduler(
            lambda epoch: 0.001 * np.exp(-0.1 * epoch)
        ),
        'Cosine Annealing': keras.callbacks.LearningRateScheduler(
            lambda epoch: 0.001 * (1 + np.cos(np.pi * epoch / 50)) / 2
        ),
        'Reduce on Plateau': keras.callbacks.ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.2,
            patience=5,
            min_lr=0.00001
        )
    }

    for name, lr_callback in techniques.items():
        print(f"\nTraining with {name} Learning Rate Scheduling")
        model = create_model()
        model.compile(
            optimizer='adam',
            loss='categorical_crossentropy',
            metrics=['accuracy']
        )
        history = model.fit(
            x_train, y_train,
            epochs=20,
            batch_size=64,
            validation_split=0.2,
            callbacks=[lr_callback],
            verbose=0
        )
        test_loss, test_accuracy = model.evaluate(x_test, y_test, verbose=0)
        print(f"Test Accuracy: {test_accuracy*100:.2f}%")
        results[name] = {
            'history': history.history,
            'test_accuracy': test_accuracy
        }

    plt.figure(figsize=(15, 10))
    plt.subplot(2, 2, 1)
    for name, result in results.items():
        plt.plot(result['history']['val_accuracy'], label=name)
    plt.title('Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.subplot(2, 2, 2)
    for name, result in results.items():
        plt.plot(result['history']['val_loss'], label=name)
    plt.title('Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.subplot(2, 2, 3)
    for name, result in results.items():
        if 'lr' in result['history']:
            plt.plot(result['history']['lr'], label=name)
    plt.title('Learning Rate')
    plt.xlabel('Epochs')
    plt.ylabel('Learning Rate')
    plt.legend()
    plt.subplot(2, 2, 4)
    techniques_list = list(results.keys())
    accuracies = [results[tech]['test_accuracy']*100 for tech in techniques_list]
    plt.bar(techniques_list, accuracies)
    plt.title('Test Accuracy by Technique')
    plt.xlabel('Learning Rate Scheduling')
    plt.ylabel('Test Accuracy (%)')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    print("\nDetailed Results:")
    for name, result in results.items():
        print(f"{name}:")
        print(f"  Test Accuracy: {result['test_accuracy']*100:.2f}%")

learning_rate_techniques()
```

```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 8)
A = np.array([0, 0, 1, 1, 0, 1, 1, 0])
B = np.array([0, 0, 0, 1, 1, 1, 0, 0])
C = np.array([0, 1, 0, 0, 0, 1, 1, 0])

def nand_gate(*args):
    return int(not all(args))

AB = np.array([nand_gate(A[i], B[i]) for i in t])
BC = np.array([nand_gate(B[i], C[i]) for i in t])
ABC_BLOW = np.array([nand_gate(A[i], 0, C[i]) for i in t])

fig, axs = plt.subplots(4, 1, figsize=(10, 6), sharex=True)

axs[0].step(t, A, where='post', color='darkred')
axs[0].step(t, B, where='post', color='darkred')
axs[0].step(t, C, where='post', color='darkred')
axs[0].set_yticks([0, 1])
axs[0].set_yticklabels(['0', '1'])
axs[0].set_ylabel('')
axs[0].set_title('A               B               C')

axs[1].step(t, AB, where='post', color='darkmagenta')
axs[1].set_yticks([0, 1])
axs[1].set_ylabel('NAND(A, B)')

axs[2].step(t, BC, where='post', color='darkorange')
axs[2].set_yticks([0, 1])
axs[2].set_ylabel('NAND(B, C)')

axs[3].step(t, ABC_BLOW, where='post', color='darkgreen')
axs[3].set_yticks([0, 1])
axs[3].set_ylabel('NAND(A, LOW, C)')

for ax in axs:
    ax.set_ylim(-0.2, 1.2)
    ax.grid(True, which='both', axis='x', linestyle='--', linewidth=0.5)

plt.tight_layout()
plt.show()
```
